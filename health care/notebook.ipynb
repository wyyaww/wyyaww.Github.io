# Analysis the data from different ways

Hospitals don't want patients staying in the hospital longer than needed so they can open up new beds, as well as not spend unnecessary money on the patients who could be home.Our health care data analyst boss wants to know what the distribution of time spent in the hospital looks like. They're also curious to know if the majority stay less than 7 days. Once patients stay over 7 days, the hospital wants to ensure these patients are very acute.

One of the best & easiest ways to show a **_distribution_** of a numerical column is by making a histogram.

```python
plt.hist(df.time_in_hospital, bins = 30, color = 'skyblue', edgecolor = 'black')
plt.show()
```

As we can see, most of people stay short than 7 days. 

Procedures are the biggest cost to a hospital. What medical specialties are doing the most number of procedures on **_average_**? 

```python
SELECT medical_specialty, ROUND(AVG(num_procedures), 1) AS avg_procedures, COUNT(*) AS patients
FROM 'diabetic_data.csv' 
GROUP BY medical_specialty
ORDER BY avg_procedures DESC
 
```

Some of these medical specialty only has 1 patiet. It's possibile. It's also may be a data entry error. I'd like to focus on the specialties with more than 50 patients. 

```python
SELECT medical_specialty, ROUND(AVG(num_procedures), 1) AS avg_procedures, COUNT(*) AS patients
FROM 'diabetic_data.csv'  
GROUP BY medical_specialty
HAVING COUNT(*) > 50
ORDER BY avg_procedures DESC
```

Let's to explore if the hospital seems to be treating patients of different races differently, specifically with the number of lab procedures done.

```python
SELECT race , AVG(num_lab_procedures) AS num_avg_lab_procedures 
FROM 'diabetic_data.csv' 
GROUP BY race
```

I don't see gaps between races from here. 

To give some insight on how the number of lab procedures might correlate to the number of days in hospital. They want to know do those who get a lot of lab procedures, stay in the hospital longer? 

We might want to categorize the patient as few procedures, average procedures, and many procedures but right now we only have the actual raw number of lab procedures. But we can group each patient into one of these categories with the CASE WHEN statement.

```python
SELECT AVG(time_in_hospital) AS avg_time,
CASE 
    WHEN num_lab_procedures >= 0 AND num_lab_procedures < 33 THEN 'few'
    WHEN num_lab_procedures >= 33 AND num_lab_procedures < 66 THEN 'average'
    ELSE 'many'
END AS procedure_frequency
FROM 'diabetic_data.csv'
GROUP BY procedure_frequency
ORDER BY avg_time DESC;
```

It looks there is a correlation between these. The longer someone is at the hospital the more lab procedures they have, or vice-versa.
